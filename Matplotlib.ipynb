```python
# Plotagem de Objeto Orientado (OO-STYLE)
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots() #Cria a figura e os eixos
ax.plot(x, x, label='linear')#plota alguns dados nos eixos
ax.plot(x, x**2, label='quadratic') #plota mais dados nos eixos
ax.plot(x, x**3, label='cubic') #...e mais alguns dados
ax.set_xlabel('x label') #Adiciona 'x label' aos eixos
ax.set_ylabel('y label') #Adiciona 'y label' aos eixos
ax.set_title("Simple Plot")
ax.legend()
```

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4,])
plt.ylabel('Alguns Números') # Nomeando o eixo X
plt.xlabel('Mais Alguns Números') # Nomeando o eixo y
plt.title('Exemplo Rápido') # Dando um Titulo para o Gráfico
plt.show()
```

Pode estar a perguntar-se porque é que o eixo x varia entre 0-3 e o eixo y entre 1-4. Se você fornecer uma única lista ou array para o comando plot(), matplotlib assume que é uma sequência de valores y, e gera automaticamente os valores x para você. Uma vez que os intervalos python começam com 0, o vector padrão x tem o mesmo comprimento que y mas começa com 0. Assim, os dados x são [0, 1, 2, 3].

```python
#Para plotar o eixo x versus o eixo y você precisa utilizar o comando
plt.plot([1,2,3,4], [2,4,1,2]) #plt.plot([x,y])
```

Para cada x, y par de argumentos, existe um terceiro argumento opcional que é a string de formato que indica a cor e o tipo de linha da parcela. As letras e símbolos da cadeia de formato são de MATLAB, e você concatenou uma cadeia de cor com uma cadeia de estilo de linha. A cadeia de formatação padrão é "b-", que é uma linha azul sólida. Por exemplo, para traçar o gráfico acima com círculos vermelhos, você emitiria

```python
plt.plot([1,2,3,4],[2,4,1,2], 'ro') # No comando 'ro': r = A cor 
plt.axis([0,5,0,5])                 # e o = O formato 
plt.show()
```

    <h2>Plotagem com cadeias de palavras-chave<a class="headerlink" href="#plotting-with-keyword-strings" title="Permalink to this headline">¶</a></h2>
    <p>aqui estão alguns casos em que você tem dados num formato que lhe permite
aceder a variáveis particulares com cadeias de caracteres. Por exemplo, com
<a class="reference external" href="https://docs.scipy.org/doc/numpy/reference/generated/numpy.recarray.html#numpy.recarray" title="(in NumPy v1.17)"><code class="xref py py-class docutils literal notranslate"><span class="pre">numpy.recarray</span></code></a> or <a class="reference external" href="https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html#pandas.DataFrame" title="(in pandas v1.0.3)"><code class="xref py py-class docutils literal notranslate"><span class="pre">pandas.DataFrame</span></code></a>.</p>
    </html>
    

```python
data = {'a' : np.arange(50),
        'c' : np.random.randint(0,50,50),
        'd' : np.random.randn(50)}

data['b']  = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a','b', c='c', s='d', data=data)
plt.xlabel('entrada a')
plt.ylabel('entrada b')
plt.show()
```

<html>
    <h2>Plotando com variaveis categóricas<a class="headerlink" href="#plotting-with-categorical-variables" title="Permalink to this headline">¶</a></h2>
    <p>Também é possível criar um gráfico utilizando variáveis categóricas.
O Matplotlib permite passar as variáveis categóricas directamente para
muitas funções de conspiração. Por exemplo:</p>
    </html>

```python
names = ['grupo_A','grupo_B','grupo_C']
values = [1,10,100]


plt.figure(figsize=(10,5))

plt.subplot(131)
plt.bar(names, values) #Plotando barras
plt.subplot(132)
plt.scatter(names, values) # plotando pontos
plt.subplot(133)
plt.plot(names, values) #plotando linhas
plt.suptitle('Plotagem Categórica')
plt.show()
```

    <h2>Trabalhando com multiplas figuras e eixos</h2>
    </html>

```python
def f(t):
    return np.exp( -t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo',t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

<html>
    <h2> Trabalhando com texto</h2>
    <p>O comando <a class="reference internal" href="../../api/_as_gen/matplotlib.pyplot.text.html#matplotlib.pyplot.text" title="matplotlib.pyplot.text"><code class="xref py py-func docutils literal notranslate"><span class="pre">text()</span></code></a> pode ser usado para adicionar um texto em uma posição arbitraria, e os comandos <a class="reference internal" href="../../api/_as_gen/matplotlib.pyplot.xlabel.html#matplotlib.pyplot.xlabel" title="matplotlib.pyplot.xlabel"><code class="xref py py-func docutils literal notranslate"><span class="pre">xlabel()</span></code></a>,
<a class="reference internal" href="../../api/_as_gen/matplotlib.pyplot.ylabel.html#matplotlib.pyplot.ylabel" title="matplotlib.pyplot.ylabel"><code class="xref py py-func docutils literal notranslate"><span class="pre">ylabel()</span></code></a> e <a class="reference internal" href="../../api/_as_gen/matplotlib.pyplot.title.html#matplotlib.pyplot.title" title="matplotlib.pyplot.title"><code class="xref py py-func docutils literal notranslate"><span class="pre">title()</span></code></a>
são utilizados para colocar textos em posições definidas (veja <a class="reference internal" href="../text/text_intro.html"><span class="doc">Text in Matplotlib Plots</span></a>
para mais detalhes)</p>
    </html>

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='c', alpha=0.75)


plt.xlabel('Inteligengia')
plt.ylabel('Probabilidades')
plt.title('Histograma do QI')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
